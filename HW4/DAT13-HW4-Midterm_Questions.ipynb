**Data Set Information:**

The "spam" concept is diverse: advertisements for products/web sites, make money fast schemes, chain letters, pornography...

**Attribute Information:**

The last column of 'spambase.data' denotes whether the e-mail was considered spam (1) or not (0), i.e. unsolicited commercial e-mail. Most of the attributes indicate whether a particular word or character was frequently occuring in the e-mail. The run-length attributes (55-57) measure the length of sequences of consecutive capital letters.

###Definitions of the attributes:

**48 continuous real [0,100] attributes of type word_freq_WORD**

= percentage of words in the e-mail that match WORD, i.e. 100 * (number of times the WORD appears in the e-mail) / total number of words in e-mail. A "word" in this case is any string of alphanumeric characters bounded by non-alphanumeric characters or end-of-string.

**6 continuous real [0,100] attributes of type char_freq_CHAR**

= percentage of characters in the e-mail that match CHAR, i.e. 100 * (number of CHAR occurences) / total characters in e-mail

**1 continuous real [1,...] attribute of type capital_run_length_average**

= average length of uninterrupted sequences of capital letters

**1 continuous integer [1,...] attribute of type capital_run_length_longest**

= length of longest uninterrupted sequence of capital letters

**1 continuous integer [1,...] attribute of type capital_run_length_total**

= sum of length of uninterrupted sequences of capital letters = total number of capital letters in the e-mail

**1 nominal {0,1} class attribute of type spam**

= denotes whether the e-mail was considered spam (1) or not (0), i.e. unsolicited commercial e-mail.

###Exercise 1: Explore the dataset and display some visualizations showing how the variables relate to each other

```python
def plot_scatter(df, x_features, y_features):    
    fig = plt.figure(figsize=(21,12))
    fig.suptitle('Comparison of features', size=32)
    sns.set_style('darkgrid')
    position = 221
    for idx in range(len(x_features)):        
        x_feature = x_features[idx]
        y_feature = y_features[idx]
        ax = fig.add_subplot(position)
        color_vector = ['#2ecc71','#9b59b6']

        for is_spam in [0,1]:         
            spam_filter = df['is_spam'] == is_spam
            x_vector = df[x_feature][spam_filter].values    
            y_vector = df[y_feature][spam_filter].values
            ax.scatter(x_vector, y_vector, color=color_vector[is_spam])

        ax.set_title(x_feature + " vs. "+ y_feature)
        ax.set_ylabel(y_feature)
        ax.set_xlabel(x_feature)
        ax.legend(["Not Spam","Spam"], fontsize=15, loc=1)
        plt.xlim(df[x_feature].min(), df[x_feature].max())
        plt.ylim(df[y_feature].min(), df[y_feature].max())
        position = position + 1
```

```python
x_features = ["char_freq_;", "capital_run_length_total",   "word_freq_email",  "word_freq_george"]
y_features = ["char_freq_$", "capital_run_length_longest", "word_freq_credit", "char_freq_["]
plot_scatter(spam_data,x_features,y_features)
```

###Exercise 2: Build a simple logistic regression and visualize it
use the variable "capital_run_length_longest" to predict "is_spam" How accurate is this single feature?

```python
from sklearn.cross_validation import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

single_feature = pd.DataFrame(spam_data.capital_run_length_longest)
target = spam_data.is_spam
features_train, features_test, target_train, target_test = train_test_split(single_feature, target, test_size=0.20, random_state=0)

# train logistic regression, evaluate on test
lr = LogisticRegression()
lr.fit(features_train, target_train)
target_predicted = lr.predict(features_test)
print(classification_report(target_test, target_predicted,
                           target_names=['not spam', 'spam']))

pl = pd.crosstab(single_feature, target.astype(bool)).plot(kind='bar')
pl.tick_params(axis='both', which='major', labelsize=10)
pl.set_xticks([1])
pl.axes.set_xlabel("capital_run_length_longest")
```


###Exercise 7: Plot the ROC Curve for the logistic regression you chose¶

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import auc

def plot_roc_curve(target_test, target_predicted_proba,threshold=0.5):
    fpr, tpr, thresholds = roc_curve(target_test, target_predicted_proba[:, 1])

    roc_auc = auc(fpr, tpr)
    # Plot ROC curve
    plt.plot(fpr, tpr, label='ROC Area = %0.3f with threshold = %0.1f' % (roc_auc, threshold))
    plt.plot([0, 1], [0, 1], 'k--')  # random predictions curve
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate or (1 - Specifity)')
    plt.ylabel('True Positive Rate or (Sensitivity)')
    plt.title('ROC')
    plt.legend(loc="lower right")    
```

```python
X = spam_data.drop("is_spam",1)
y = spam_data.is_spam
features_train, features_test, target_train, target_test = train_test_split(X, y, random_state=12)
lr = LogisticRegression()
lr.fit(features_train,target_train)
proba = lr.predict_proba(features_test)
print proba[:10]
plot_roc_curve(target_test, proba)
```

###Exercise 8: Demonstrate how the accurary of your predictions changes when you set your threshold to levels other than 50%

```python
class MyLogisticRegression(LogisticRegression):
    
    def predict_proba_with_threshold(self, X, threshold=0.5):
        probs = self.predict_proba(X)
        predictions = []
        for prob in probs:
            vals = []
            prob_0 = prob[0]
            prob_1 = prob[1]
            swap = False
            
            if threshold >= 0.5:
                if(prob_0 > prob_1) and (prob_0 < threshold):
                    swap = True
                if(prob_1 > prob_0) and (prob_1 < threshold):
                    swap = True
            else:
                if(prob_0 < prob_1) and (prob_0 > threshold):
                    swap = True                   
                if(prob_1 < prob_0) and (prob_1 > threshold):
                    swap = True
            
            if (swap==True):
                vals = np.array([prob_1, prob_0])
            else:
                vals = np.array([prob_0, prob_1])
                
            predictions.append(vals)

        return np.array(predictions)
```

```python
plt.figure(figsize=(15,12))
plt.tick_params(axis='both', which='major', labelsize=10)

X = spam_data.drop("is_spam",1)
y = spam_data.is_spam
features_train, features_test, target_train, target_test = train_test_split(X, y, random_state=12)
mlr = MyLogisticRegression()
mlr.fit(features_train, target_train)

for threshold in np.arange(0,1.1,0.1):
    mproba = mlr.predict_proba_with_threshold(features_test,threshold)
    plot_roc_curve(target_test, mproba, threshold)
```

As shown in the graph above, moving the threshold away from 50% increases the false positive rates. For our binary classifier, a 50% threshold gets us the closest to the 100% true positive rate.  
In rare cases you may want to move this percentage.  For instance when using the model to predict a terminal disease and false positives cannot be tolerated (in those instances an ROC curve is not a good evaluator of the model because moving the threshold higher than 50% still resulted in an increase in the false positive rate). 
<p>In most situations, it's not a good idea to move the threshold.</p>
